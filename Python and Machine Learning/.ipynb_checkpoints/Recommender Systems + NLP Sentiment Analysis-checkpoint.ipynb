## user-user analysis

```python
# The function below identifies the top 30 similar users.

def find_n_neighbours(df, n):
    order = np.argsort(df.values, axis=1)[:, :n]
    df = df.apply(lambda x: pd.Series(x.sort_values(ascending=False).iloc[:n].index, index=[f"top{i}" for i in range(1, n+1)]), axis=1)
    return df
```

### User-item Analysis

##### The function below will provide items that the chosen user users have rated (quality feature).

```python
def get_user_similar_items(user1, user2):
    items_user1_rated = df_rec_items[df_rec_items["user_id"] == user1]
    items_user2_rated = df_rec_items[df_rec_items["user_id"] == user2]
    common_items = items_user1_rated.merge(items_user2_rated, on="item_id", how="inner", suffixes=(f"_{user1}", f"_{user2}"))
    return common_items
```

##### The user item function provides a score of how much the chosen user likes the chosen item (a higher score is better)

```python
# user_id from mean rating variable needs to be changed to integer for the function purposes
mean_rating["user_id"] =  mean_rating["user_id"].astype(int)
```

```python
def user_item_score(user, item):
    user_values = top_30_users[top_30_users.index == user].values
    user_values_squeezed = user_values.squeeze().tolist()
    item = items_df.loc[:, item]
    item_in_list = item[item.index.isin(user_values_squeezed)]
    item_in_list = item_in_list[item_in_list.notnull()]
    
    mean_usr_rating = mean_rating.loc[mean_rating['user_id'] == user, 'quality'].values[0]
    index = item_in_list.index.values.squeeze().tolist()
    corr = similarity_df.loc[user, index]
    
    user_item_df = pd.concat([item_in_list, corr], axis=1)
    user_item_df.columns = ["mean_score", "correlation"]
    user_item_df["score"] = user_item_df.apply(lambda x: x['mean_score'] * x['correlation'], axis=1)
    final_score = user_item_df["score"].sum() / user_item_df["correlation"].sum()
    return final_score
    
```

##### The function did not found recommendations in top 30 users

```python
def get_recommendations(user):
    items_seen_by_user = pivot_df.columns[pivot_df[pivot_df.index == user].notna().any()].tolist()
    top_users = top_30_users[top_30_users.index == user].values.squeeze().tolist()
    top_users_items = df_str[df_str.index.isin(top_users)]
    combination_df = ','.join(top_users_items.values)
    print(combination_df)
    
    items_seen_by_similar_users = combination_df.split(',')
    #potential_recommendations = list(set(items_seen_by_similar_users) - set(list(map(str, items_seen_by_user))))
    potential_recommendations = list(set(items_seen_by_similar_users) - set(map(str, items_seen_by_user)))
    potential_recommendations = list(map(int, potential_recommendations))
    
    scores = []
    for item in potential_recommendations:
        scores.append(user_item_score(user, item))
        
    recommendation_df = pd.DataFrame({'item_id':potential_recommendations, 'score': scores})
    recommendation_df.sort_values(by='score', ascending=False, inplace=True)
    #recommendation_df = recommendation_df.merge(df_rec_items, how="inner", on="item_id")
    return recommendation_df
```

##### The function below computes the weighted quality score per item

```python
def weighted_rating(x, m=m, C=C):
    v = x['quality_count']
    R = x['quality_mean']
    # Calculation based on the IMDB formula
    return round((v / (v + m) * R) + (m / (m + v) * C),2)
```

##### Feature creation of weight rating quality score 

```python
q_df_ratings_item['weight_rating_quality_score'] = q_df_ratings_item.apply(weighted_rating, axis = 1)
```

### Review based Recomendations

```python
# Function that takes in item title as input and outputs most similar items
def get_recommendations(title, cosine_sim=cosine_sim):
    # Get the index of the item that matches the title
    idx = indices[title]

    # Get the pairwsie similarity scores of all items with that item
    sim_scores = list(enumerate(cosine_sim[idx]))

    # Sort the items based on the similarity scores
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)

    # Get the scores of the 10 most similar items
    sim_scores = sim_scores[1:11]

    # Get the item indices
    item_indices = [i[0] for i in sim_scores]

    # Return the top 10 most similar items
    return q_df_ratings_item['user_item_category_id'].iloc[item_indices]
```

##### The function below would convert sentiment scores to a more consistent form which is positive, neutral or negative.


```python
def OutcomeAnalysis(i):
    if i > 0:
        return "Positive"
    elif i == 0:
        return "Neutral"
    else:
        return "Negative"
```

```python
q_df_ratings_item["textBlob_outcome_analysis"] = q_df_ratings_item["TextBlob_sentiment"].apply(OutcomeAnalysis )
```
